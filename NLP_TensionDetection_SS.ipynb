```python
tension_tokens=[(nltk.word_tokenize(t),'Tension') for t in tension_sent_tokens]
no_tension_tokens=[(nltk.word_tokenize(t),'No Tension') for t in no_tension_sent_tokens]
```

```python
documents = tension_tokens+no_tension_tokens
```

```python
import random
random.seed(4)
random.shuffle(documents)
```

```python
documents[0]
```

```python
all_words_list = [word for (sent,cat) in documents for word in sent]
all_words = nltk.FreqDist(all_words_list)
word_items = all_words.most_common(2000)
word_features = [word for (word,count) in word_items]
word_features[200]
```

```python
def document_features(document, word_features):
    document_words = set(document)
    features = {}
    for word in word_features:
        features['contains({})'.format(word)] = (word in document_words)
    return features
```

```python
featuresets = [(document_features(d, word_features), c) for (d, c) in documents]
featuresets[0]
```

```python
len(featuresets)
```

```python
train_set, test_set = featuresets[:3328], featuresets[3328:]
classifier = nltk.NaiveBayesClassifier.train(train_set)
nltk.classify.accuracy(classifier, test_set)
```

```python
import collections
from nltk.metrics import *
def metrics(classifier, test):
    actual_sets = collections.defaultdict(set)
    predicted_sets = collections.defaultdict(set)
    for i, (feature, label) in enumerate(test):
        actual_sets[label].add(i)
        predicted = classifier.classify(feature)
        predicted_sets[predicted].add(i)
    print('Precision for +ve :', precision(actual_sets['Tension'], predicted_sets['Tension']))
    print('Recall for +ve ', recall(actual_sets['Tension'], predicted_sets['Tension']))
    print('F-Measure for +ve ', f_measure(actual_sets['Tension'], predicted_sets['Tension']))
    print('Precision for -ve ', precision(actual_sets['No Tension'], predicted_sets['No Tension']))
    print ('Recall for -ve ', recall(actual_sets['No Tension'], predicted_sets['No Tension']))
    print ('F-Measure for -ve ', f_measure(actual_sets['No Tension'], predicted_sets['No Tension']))
```

```python
metrics(classifier,test_set)
```

```python
from nltk.classify import SklearnClassifier
from sklearn.svm import SVC

svm_classifier = SklearnClassifier(SVC()).train(train_set)
print ('accuracy:', nltk.classify.util.accuracy(svm_classifier, test_set))
```

```python
from nltk.classify import SklearnClassifier

DT_classifier = nltk.classify.DecisionTreeClassifier.train(train_set)
print ('accuracy:', nltk.classify.util.accuracy(DT_classifier, test_set))
```
